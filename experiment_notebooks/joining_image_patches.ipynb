```python
import os
import json
import glob


patch_file = "/workspace/Dataset-Curation/curated_data/image/blue_patches"

files=[]
large_dict={}
```

```python
for filepath in glob.iglob(f"{patch_file}/*.json"):
    image_id  = os.path.splitext(os.path.basename(filepath))[0].split("_")[0]
    
   # print(image_id)

    with open(filepath, 'r') as f:
        content = json.load(f)
    #print(content)

    large_dict[image_id]=content
```
