```python
import json
import os
import sys
from PIL import Image
import shutil
```

```python
def is_image_in_file(image_dir,image_id):

    extentions=[".jpg", ".jpeg", ".png"]

    for ext in extentions:
        image_path = os.path.join(image_dir, f"{image_id}{ext}")
        if os.path.exists(image_path):
            return True, image_path
        
    return False, None
```

```python
def get_image_dimensions(image_path):
    with Image.open(image_path) as img:
        return img.size 
```

```python
def save_image_filtered(img_path,image_id,output_dir):
    
    os.makedirs(output_dir, exist_ok=True)
    filename = f"{image_id}.jpg"
    output_path = os.path.join(output_dir, filename)  
    shutil.copy(img_path, output_path)
    
```

```python
def get_concept(image_ann,image_dir,output_dir,concept='red',threshold=0.005):

    
    image_id=image_ann.get("image_id")
    #print("image id:",image_id)
    result = is_image_in_file(image_dir, image_id)
    if not result:
        return False
    is_found, img_path = result
    

    if is_found:
        image_width, image_height= get_image_dimensions(img_path)
        image_size=image_width*image_height
    else:
        return False
    
    

    for obj in image_ann.get("attributes", []):
            
            obj_attributes = obj.get("attributes")
            #print("object att: ", obj_attributes)
            if not obj_attributes:
                continue

            obj_attributes_lower = [attr.lower() for attr in obj_attributes]

            if concept.lower() not in obj_attributes_lower :
                continue
            
            print(f"found {concept} in image {image_id}")
            obj_width, obj_height =  obj.get("w", 0), obj.get("h", 0)
            obj_area=obj_width*obj_height

            if (obj_area / image_size) > threshold:
                 
                save_image_filtered(img_path,image_id,output_dir)
                
                return True
            
    return False
                
    
            
    
                 
                 
    



```

```python
def process_all_images(json_path, image_dir, output_dir, concept='red', threshold=0.005,size=50):
    count = 0
    with open(json_path, 'r') as f:
        data = json.load(f)
        for image_ann in data:
            if count >= size: 
                break
            saved=get_concept(image_ann, image_dir, output_dir, concept, threshold)
            if saved:
                count+=1
```
